```python
import numpy as np
from collections import Counter

def replace_delimiters_with_mean(arr, threshold=0.05):
    # Convert to numpy array for easier manipulation
    arr = np.array(arr)

    # Calculate the frequency of each element
    counts = Counter(arr)

    # Determine the threshold count based on the given threshold percentage
    total_elements = len(arr)
    threshold_count = total_elements * threshold

    # Identify delimiters based on the threshold count
    delimiters = [key for key, count in counts.items() if count < threshold_count]

    # Mask to identify delimiters
    mask = np.isin(arr, delimiters)

    # Calculate mean of non-delimiters
    non_delimiters = arr[~mask]
    mean_value = np.mean(non_delimiters)

    # Replace delimiters with the mean value
    arr[mask] = mean_value

    return arr, delimiters, mean_value

# Example usage
array = [1, 2, -1, 4, -1, 6, 7, -1]
result, identified_delimiters, mean_value = replace_delimiters_with_mean(array)

print(f"Resulting array: {result}")
print(f"Identified delimiters: {identified_delimiters}")
print(f"Calculated mean value: {mean_value}")
```
